1)Top 10 selling products

```python
top_selling_products=df.groupby('Product Name').agg({'Sales': 'sum', 'Quantity': 'sum'})
top_10_selling_products=top_selling_products.head(10)
```

```python
top_10_selling_products
```

2)Profit Margins

```python
profit_by_produt=df.groupby('Product Name').agg({'Profit':'sum'})
profit_by_produt.sort_values(by='Profit', ascending=False).head()
```

```python
profit_by_category=df.groupby('Category').agg({'Profit':'sum'})
profit_by_category.sort_values(by='Profit', ascending=False).head()
```

```python
profit_by_Region=df.groupby('Region').agg({'Profit':'sum'})
profit_by_Region.sort_values(by='Profit', ascending=False).head()
```

3)Loss-Making

```python
loss_making_products=df.query('Profit<0').groupby('Product Name').agg({'Profit':'sum'})
loss_making_products.sort_values(by='Profit',ascending=False).head()
```

```python
loss_making_categories=df.query('Profit<0').groupby('Category').agg({'Profit':'sum'})
loss_making_categories.sort_values(by='Profit',ascending=False).head()
```

4) Customer Segmentation - Analyze Segment for sales and profit:

```python
segment_analysis = df.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'})
segment_analysis.head()
```

```python
customer_loyalty_value = df.groupby('Customer ID').agg({'Order ID': pd.Series.nunique, 'Sales': 'sum'})
customer_loyalty_value.sort_values(by='Sales', ascending=False).head()
```

5)Total Sales in city,state, region

```python
sales_by_city = df.groupby('City').agg({'Sales': 'sum'})
sales_by_city.sort_values(by='Sales', ascending=False).head()
```

sales_by_state = df.groupby('State').agg({'Sales': 'sum'})
sales_by_state.sort_values(by='Sales', ascending=False).head()

```python
sales_by_region = df.groupby('Region').agg({'Sales': 'sum'})
sales_by_region.sort_values(by='Sales', ascending=False).head()
```

6)Total Profit in regions

```python
regional_profit = df.groupby('Region').agg({'Profit': 'sum'})
regional_profit.sort_values(by='Profit', ascending=False)
```

7)Shipping Efficiency

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df['Shipping Duration'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
ship_mode = df.groupby('Ship Mode').agg({'Shipping Duration': 'mean', 'Sales': 'sum', 'Returns': 'count'})
ship_mode
```

8)Customer satisfaction based on sales and returns per shipping mode

```python
ship_mode_satisfaction = df.groupby('Ship Mode').agg({'Sales': 'sum', 'Returns': 'count'})
ship_mode_satisfaction.head()
```

9)Return Analysis 

```python
returns_by_product = df[df['Returns'] == 'Yes'].groupby('Product Name').agg({'Returns': 'count'})
returns_by_product.sort_values(by='Returns', ascending=False).head()
```

```python
returns_by_category = df[df['Returns'] == 'Yes'].groupby('Category').agg({'Returns': 'count'})
returns_by_category.sort_values(by='Returns', ascending=False).head()
```

10)Payment Mode Preferences

```python
payment_mode_analysis = df.groupby('Payment Mode').agg({'Sales': 'sum', 'Returns': 'count'})
payment_mode_analysis
```

11)Trends by month and year

```python
df['Order Year'] = pd.to_datetime(df['Order Date']).dt.year
df['Order Month'] = pd.to_datetime(df['Order Date']).dt.month
```

```python
monthly_sales_trend = df.groupby(['Order Year', 'Order Month']).agg({'Sales': 'sum', 'Profit': 'sum'})
monthly_sales_trend
```
